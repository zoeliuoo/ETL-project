### Import pandas

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-played_video_games_by_player_count#cite_note-77'
```

### Read tables  

```python
tables = pd.read_html(url)
df1 = tables[1]
df1
```

### Extract a list of numbers

```python
number = []
for i in df1['Number']:
    if 'million' in i:
        number.append(i.split('million')[0])
    elif 'billion' in i:
        n = i.split('billion')[0]
        a = pd.to_numeric(n)*1000
        number.append(a)
    else:
        print("NA")
number = pd.to_numeric(number)
number
```

### Number of players (in millions) of most popular games according to wiki

```python
df_Game = pd.DataFrame(df1['Game'])
df_number = pd.DataFrame(number)

df1 = pd.concat([df_Game,df_number],axis=1)
df1.columns = ['Game','million_played']

df1.to_csv('wiki_most_played.csv')
df1
```
