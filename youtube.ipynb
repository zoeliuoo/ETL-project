### Import libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
from splinter import Browser
```

### Create path, define URL and ask browser to visit

```python
driverPath = !where chromedriver
executable_path = {'executable_path': driverPath[0]}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = "https://www.youtube.com/gaming/games"
browser.visit(url)
```

### Web scraping to get a list of most viewed live games

```python
html = browser.html
soup = bs(html, 'html.parser')
```

```python
contents = soup.find_all('a', class_ = 'yt-simple-endpoint style-scope ytd-game-details-renderer' )
print(contents)
```

```python
title_lst = []
view_lst = []

for content in contents:
    title = content.find(attrs={'id':'title'})
    view = content.find(attrs={'id':'live-viewers-count'})
    title_lst.append(title.text)
    view_lst.append(view.text)
    
print(title_lst)
print(view_lst)
```

### Extract a list of numbers and put together a dataframe

```python
value = []
for v in view_lst:
    if 'K' in v:
        value.append(v.split('K')[0])
value = pd.to_numeric(value)
value
```

### Number of viewers (in thousands) of live video games on you tube

```python
df = pd.concat([pd.DataFrame(title_lst),pd.DataFrame(value)],axis=1)
df.columns = ['Game','K_views']
df = df.dropna(axis=0)
df.to_csv('youtubelive.csv')
df
```
